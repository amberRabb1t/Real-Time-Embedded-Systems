```python
import csv
import os
from collections import Counter
import pandas as pd
from IPython.display import display
```

```python
subscriptions = ["BTC-USDT", "ADA-USDT", "ETH-USDT", "DOGE-USDT", "XRP-USDT", "SOL-USDT", "LTC-USDT", "BNB-USDT"]
extension = ".txt"
df = []

for label in subscriptions:

    pcc_data_label = []

    pcc_path = os.path.join('generated-files', 'pearson-correlation-coefficients', label + extension)

    # Read and convert Pearson Correlation Coefficient data
    with open(pcc_path, 'r') as pcc_file:
        pcc_reader = csv.reader(pcc_file)
        next(pcc_reader, None) # Skip header
        firstRow = next(pcc_reader, None)

        for row in pcc_reader:
            pcc_data_label.append(row[0])
    
    df.append(pd.DataFrame.from_records(list(dict(Counter(pcc_data_label)).items()), columns=['Predicted -->', label]))

finaldf = pd.merge(df[0], df[1], on='Predicted -->', how='outer')
finaldf = pd.merge(finaldf, df[2], on='Predicted -->', how='outer')
finaldf = pd.merge(finaldf, df[3], on='Predicted -->', how='outer')
finaldf = pd.merge(finaldf, df[4], on='Predicted -->', how='outer')
finaldf = pd.merge(finaldf, df[5], on='Predicted -->', how='outer')
finaldf = pd.merge(finaldf, df[6], on='Predicted -->', how='outer')
finaldf = pd.merge(finaldf, df[7], on='Predicted -->', how='outer')
display(finaldf)
```
